```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules

```

```python
data = pd.read_csv("Destination.csv")
```

```python
data.head()
```

```python
data.dtypes
```

```python
data = data.drop("Destinations",axis=1)
```

```python
data.head()
```

```python
# apply apriori algorithm
frequent_itemsets  = apriori(data, min_support=0.25,use_colnames=True)
```

```python
frequent_itemsets
```

```python
type(frequent_itemsets)
```

```python
frequent_itemsets.sort_values('support',ascending=False)
```

```python
rules = association_rules(frequent_itemsets, metric = 'confidence',min_threshold=0.4)
```

```python
rules
# lift of 1.0 means independent.  lift >1.0 means there is some correlation.  lift < 1 means substitute products
```
